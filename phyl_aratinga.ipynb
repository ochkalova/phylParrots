```python
#parse multifasta or fasta files

fasta = {} 
for id_ in ids:
    fasta_file = '/mnt/projects/sochkalova/Aratinga/phyl/%s.fasta' % (id_)
    header = id_ 
    seq = []
    with open(fasta_file) as fh: 
        for line in fh:         
            line = line.strip()         
            if line.startswith(">"):
                 continue     
            seq.append(line)             
        fasta[header] = "".join(seq)     


fasta_file = '/mnt/projects/sochkalova/Aratinga/phyl/Nine_Aratinga_aligned.mfa'          
header = None                             
with open(fasta_file) as fh:              
    for i, line in enumerate(fh):         
        line = line.strip()               
        if line.startswith(">"):          
            if header:                    
                fasta[header] = "".join(seq) 
            header = line[1:]             
            seq = []                      
        else:
            seq.append(line)              
    if header:                            
        fasta[header] = "".join(seq)  
print(*fasta)
```

```python
# make a dict with gene sequencies as items and spicies+gene name as keys

seq_frag = {}
for id_, diap in data_ncbi.items():
    start, end = diap.replace('(', '').replace(')', '').split()
    seq_frag[id_] = ''.join(fasta[id_.split()[0]][int(start):int(end)+1]).upper()
    lenght = int(end) - int(start)
    print(id_, lenght)
print(seq_frag, sep='\n')
```
